```python
review = df2["last_review"]
date = []
year = []
review.fillna("N/A")

for i in range (0, len(review)):
    date.append(review[i])
    
for i in date:
    if i != "N/A":
        year.append(str(i)[:4])
    else:
        year.append("None")

year
```
