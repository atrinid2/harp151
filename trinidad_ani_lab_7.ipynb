```python
import requests
import json
import csv
import pandas as pd
```

```python
lat = "42.098701"
lon = "-75.912537"

csv_file = open("trindad_ani_lab_7.csv", "w", newline = "", encoding="utf-8")
csv_writer = csv.writer(csv_file)
csv_writer.writerow(["Temperature", "Day", "Description"])
url = f"https://api.weather.gov/points/{lat},{lon}"

weather = requests.get(url)
file = weather.json()
forecast_file = file["properties"]
forecast = forecast_file['forecast']

forecast

new_url = forecast
weather_fr = requests.get(new_url)
file2 = weather_fr.json()
file2

one = file2["properties"]
two = one["periods"]

for i in two:
    csv_writer.writerow([i["temperature"], i["name"], i["detailedForecast"]])
    print(i["temperature"])
    print(i["name"])
    print(i["detailedForecast"])
    
csv_file.close()
```

```python
#Question 2
#LongWeekend

#all 7 of the calls are used by placing the users desired infromation in the API's url and then we use requests to display it in juypter
#each "Section" represents a key in the whole dictionary that is the API, and each call acesses a certian, second key within the section

#This one checks for long weekends in a country, the url retrieves an item in a big "database"/dictionary. In this case, it's looking for the longweekend key, and is further specified by our inputted year and country
country = "MA" #I used the country code for morocco
year = "2023"
url = f"https://date.nager.at/api/v3/LongWeekend/{year}/{country}"
data = requests.get(url)

json_file = data.json()
json_file
```

```python
#Country

#When requested, this url just returns a collection of all avialable countries thats able to be used with this API.
#there is no user input here
url1 = "https://date.nager.at/api/v3/AvailableCountries"
data = requests.get(url1)

json_file = data.json()
json_file
```

```python
#this call is used to find a specific country info in the API. Here, i spliced Morroco's code where the country should be in the url to get the specific info from the dictionary.

url2 = f"https://date.nager.at/api/v3/CountryInfo/{country}"
data = requests.get(url2)

json_file = data.json()
json_file
```

```python
#PublicHoliday

#this call is used to find public holidays, their datesm and their information in the API. Here, i spliced Morroco's code where the country should be in the url to get the specific info from the dictionary, aswell as the year.


url3 = f"https://date.nager.at/api/v3/PublicHolidays/{year}/{country}"
data = requests.get(url3)

json_file = data.json()
json_file
```

```python
#this call is a little more unique in that it automatically takes note of the date, and either gives the response 200 or 204 (204 being not a holiday, 200 being a holiday)
#I just spliced in the country code and printed what i got to find that i got a 204, meaning it is not a public holiday

url4 = f"https://date.nager.at/api/v3/IsTodayPublicHoliday/{country}"
data = requests.get(url4)

print(data)
```

```python
#this call also takes into account the date that the user is acessing it at
#here, i just put in the country code to have the program return the next holidays in order for the next year
url5 = f"https://date.nager.at/api/v3/NextPublicHolidays/{country}"

data = requests.get(url5)

json_file = data.json()
json_file
```

```python
#this call also takes in the account the date of acessesing, but returns all holidays worldwide for the next week

url6 = "https://date.nager.at/api/v3/NextPublicHolidaysWorldwide"
data = requests.get(url6)

json_file = data.json()
json_file
```
